**Prompt 1**$\,\,\,\,\,\,\,\,\,\,$Splitting Well Log Data

Generate Python code that reads CSV files from two lists, 'train_list' and 'test_list', and creates one combined pandas dataframe for each list.

Each CSV file should be read into a dataframe and concatenated into a single dataframe for train and one for test.

```python
import pandas as pd

# Example lists of CSV file paths
train_list = [r".\Well Data\15_9-F-1 B.csv", r".\Well Data\15_9-F-4.csv", r".\Well Data\15_9-F-5.csv", r".\Well Data\15_9-F-11 B.csv", r".\Well Data\15_9-F-12.csv"]
test_list = [r".\Well Data\15_9-F-14.csv", r".\Well Data\15_9-F-15 C.csv"]

def combine_csv(file_list):
    """
    Reads multiple CSV files into DataFrames and concatenates them into one.
    """
    dfs = [pd.read_csv(file) for file in file_list]
    combined_df = pd.concat(dfs, ignore_index=True)
    return combined_df

# Create combined DataFrames
train_df = combine_csv(train_list)
test_df = combine_csv(test_list)

# Optional: check results
print("Train DataFrame shape:", train_df.shape)
print("Test DataFrame shape:", test_df.shape)
```

**Prompt 2**$\,\,\,\,\,\,\,\,\,\,$Visualise Lithology Distribution

Generate Python code that takes two pandas dataframes, `train_df` and `test_df`, each containing a 'LITHOLOGY' column.

Find all unique lithology names present in either dataframe and return them as a sorted list. Ensure that the list contains no duplicates and includes all lithologies from both train and test sets.

```python
# Combine unique lithologies from both DataFrames
unique_lithologies = sorted(
    set(train_df['LITHOLOGY'].unique()) | set(test_df['LITHOLOGY'].unique())
)

print(unique_lithologies)
```

**Prompt 3**$\,\,\,\,\,\,\,\,\,\,$Visualise Lithology Distribution

Generate Python code that takes two pandas dataframes, train_df and test_df, each containing a 'LITHOLOGY' column, and a list called unique_lithologies containing all unique lithology names. Plot histograms of lithology counts for both sets using matplotlib.  Here are some additional details:

*   Use subplots to display the train and test histograms side by side.
*   Label the x-axis as 'Lithology' and the y-axis as 'Count'
*   Add titles: "Train Set Lithology Distribution" and "Test Set Lithology
*   Display the count on top of each bar in both histograms.
*   Optionally, use different colors for each lithology type.

```python
import matplotlib.pyplot as plt
import numpy as np

# Count lithologies in both sets, ensuring all unique lithologies are represented
train_counts = train_df['LITHOLOGY'].value_counts().reindex(unique_lithologies, fill_value=0)
test_counts = test_df['LITHOLOGY'].value_counts().reindex(unique_lithologies, fill_value=0)

# Assign colors (optional) â€” one color per lithology type
colors = plt.cm.tab10(np.linspace(0, 1, len(unique_lithologies)))

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

# --- Train set plot ---
axes[0].bar(unique_lithologies, train_counts.values, color=colors)
axes[0].set_title("Train Set Lithology Distribution")
axes[0].set_xlabel("Lithology")
axes[0].set_ylabel("Count")
axes[0].tick_params(axis='x', rotation=45)
# Add count labels
for i, count in enumerate(train_counts.values):
    axes[0].text(i, count + 0.5, str(count), ha='center', va='bottom')

# --- Test set plot ---
axes[1].bar(unique_lithologies, test_counts.values, color=colors)
axes[1].set_title("Test Set Lithology Distribution")
axes[1].set_xlabel("Lithology")
axes[1].tick_params(axis='x', rotation=45)
# Add count labels
for i, count in enumerate(test_counts.values):
    axes[1].text(i, count + 0.5, str(count), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
